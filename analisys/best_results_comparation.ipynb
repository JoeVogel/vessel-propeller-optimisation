```python
import pandas as pd

import csv
from os import listdir
```

```python
dir_run = '../results'
```

```python
data = []

for alg in listdir(dir_run):
    algorithm_folder = dir_run + '/' + alg
    
    splited_values = alg.split('_')
    solver = splited_values[0]
    
    if (splited_values[1] == 'mod'):
        solver = solver + '_' + 'mod'
    
    for s in listdir(algorithm_folder):
        speed_dir = algorithm_folder + '/' + s
    
        best_result_file = [filename for filename in listdir(speed_dir) if 'best_configuration' in filename]
        best_result_path = speed_dir + '/' + best_result_file[0]
        
        best_result = {
            'D':0,
            'AEdAO':0,
            'PdD':0,
            'Z':0,
            'P_B':0,
            'fitness':0,
            'solver': solver,
            'V_S': s.replace('_', '.')
        }
        
        with open(best_result_path, 'r', newline='') as file:
            reader = csv.reader(file)
            
            for row in reader:
                
                if (len(row) > 0):
                
                    match row[0]:
                        case 'D':
                            best_result['D'] = row[1]
                        case 'AEdAO':
                            best_result['AEdAO'] = row[1]
                        case 'PdD':
                            best_result['PdD'] = row[1]
                        case 'Z':
                            best_result['Z'] = row[1]
                        case 'P_B':
                            best_result['P_B'] = row[1]
                        case 'fitness':
                            best_result['fitness'] = row[1]
        
        data.append(best_result)

df = pd.DataFrame(data)
```
